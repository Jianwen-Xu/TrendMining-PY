```python
import sys; sys.path.insert(0, "..")
from config.settings import SCOPUS_API_KEY, SO_API_KEY, DATA_DIR
from src.fetchers.scopus import ScopusClient
from src.fetchers.stackoverflow import StackOverflowClient
from src.fetchers.github_trending import fetch_github_topics, fetch_github_trending
from src.utils.data_store import save, load, exists
```

```python
# Scopus — sorted by citation count for historical year spread
# Requires SCOPUS_API_KEY in .env or passed directly
client = ScopusClient(SCOPUS_API_KEY)
df_scopus = client.fetch("devops", max_results=500)
save(df_scopus, "scopus_devops", DATA_DIR)
df_scopus.head()
```

```python
# StackOverflow — API key optional (300 req/day without key, max 2500 results)
# Leave SO_API_KEY empty in .env to use unauthenticated
so_client = StackOverflowClient(SO_API_KEY)
df_so = so_client.fetch("devops")
save(df_so, "so_devops", DATA_DIR)
df_so.head()
```

```python
# GitHub Topics — keyword searchable, optional language filter
# Scrapes github.com/topics/{topic}?l={language}
df_gh_topics = fetch_github_topics(topic="devops", language="", pages=3)
save(df_gh_topics, "github_topics_devops_all", DATA_DIR)
print(f"Fetched {len(df_gh_topics)} repos")
df_gh_topics.head()
```

```python
# GitHub Trending — today's hot repos, no keyword filter
# Scrapes github.com/trending/{language}?since={period}
df_gh_trend = fetch_github_trending(language="python", period="weekly")
save(df_gh_trend, "github_trending_python_weekly", DATA_DIR)
print(f"Fetched {len(df_gh_trend)} trending repos")
df_gh_trend.head()
```
